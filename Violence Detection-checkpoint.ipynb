# Frame Extraction

```python
def extract_frames(video_path, output_dir, start_frame_number=0, frames_per_video=16):
    video_reader = cv2.VideoCapture(video_path)
    total_frames = int(video_reader.get(cv2.CAP_PROP_FRAME_COUNT))
    
    skip_frames = max(total_frames // frames_per_video, 1)

    # Extract frames
    for i in range(frames_per_video):
        video_reader.set(cv2.CAP_PROP_POS_FRAMES, i * skip_frames)
        success, frame = video_reader.read()
        if success:
            frame_path = os.path.join(output_dir, f"frame_{start_frame_number + i}.jpg")
            cv2.imwrite(frame_path, frame)
    video_reader.release()

def process_dataset(dataset_dir, output_dir, frames_per_video=16):
    overall_frame_number = 0 
    
    for class_name in os.listdir(dataset_dir):
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path): 
            print(f"Processing class: {class_name}")
            class_output_dir = os.path.join(output_dir, class_name)
            os.makedirs(class_output_dir, exist_ok=True)
            # Process each video
            for video_file in os.listdir(class_path):
                video_path = os.path.join(class_path, video_file)
                if video_file.endswith(('.mp4', '.avi', '.mkv')): 
                    video_name = os.path.splitext(video_file)[0]
                    video_output_dir = os.path.join(class_output_dir, video_name)
                    os.makedirs(video_output_dir, exist_ok=True)
                    extract_frames(video_path, video_output_dir, start_frame_number=overall_frame_number, frames_per_video=frames_per_video)
                    overall_frame_number += frames_per_video

dataset_dir = "250Dataset"
output_dir = "OP_Frames"
process_dataset(dataset_dir, output_dir, frames_per_video=16)
```

# Creating Labels for Classification
0 is for Non Violent, 1 is for Violent

```python
def generate_labels(violent_directory, output_label_file="label250.txt"):
    with open(output_label_file, 'a') as label_file:
        for class_name in os.listdir(violent_directory):
            class_path = os.path.join(violent_directory, class_name)
            if os.path.isdir(class_path):
                print(f"Processing class: {class_name}")
                # Process each video
                for video_folder in os.listdir(class_path):
                    video_folder_path = os.path.join(class_path, video_folder)
                    if os.path.isdir(video_folder_path):  
                        # Process each frame
                        for frame_file in os.listdir(video_folder_path):
                            if frame_file.endswith('.jpg'):
                                frame_path = os.path.join(video_folder_path, frame_file)
                                frame_path_fixed = frame_path.replace("\\", "/")
                                frame_path_cleaned = frame_path_fixed.replace("//", "/")
                                if class_name == "NonViolence":
                                    label_file.write(f"{frame_path_cleaned} 0\n")
                                else:
                                    label_file.write(f"{frame_path_cleaned} 1\n")

violent_directory = "OP_Frames"
generate_labels(violent_directory, output_label_file="label250.txt")
```

# Split to Train and Validation Sets
The train and validation sets are created in the ration 80:20

```python
from shutil import copy2

def split_lf(label_file, output_dir, train_ratio=0.8):
    train_images_dir = os.path.join(output_dir, 'images/train250')
    val_images_dir = os.path.join(output_dir, 'images/val250')
    train_labels_file = os.path.join(output_dir, 'train250.txt').replace("\\", "/")
    val_labels_file = os.path.join(output_dir, 'val250.txt').replace("\\", "/")
    os.makedirs(train_images_dir, exist_ok=True)
    os.makedirs(val_images_dir, exist_ok=True)
    with open(label_file, 'r') as file:
        lines = file.readlines()

    violent_samples = []
    nonviolent_samples = []
    for line in lines:
        parts = line.strip().rsplit(' ', 1)
        if len(parts) == 2:
            frame_path, label = parts
            if label == '1':
                violent_samples.append(line)
            elif label == '0':
                nonviolent_samples.append(line)

    # Shuffle samples
    random.shuffle(violent_samples)
    random.shuffle(nonviolent_samples)

    # Split each class
    violent_split_idx = int(len(violent_samples) * train_ratio)
    nonviolent_split_idx = int(len(nonviolent_samples) * train_ratio)
    train_violent = violent_samples[:violent_split_idx]
    val_violent = violent_samples[violent_split_idx:]
    train_nonviolent = nonviolent_samples[:nonviolent_split_idx]
    val_nonviolent = nonviolent_samples[nonviolent_split_idx:]
    # Combine and shuffle train and validation splits
    train_lines = train_violent + train_nonviolent
    val_lines = val_violent + val_nonviolent
    random.shuffle(train_lines)
    random.shuffle(val_lines)

    with open(train_labels_file, 'w') as train_file, open(val_labels_file, 'w') as val_file:
        for line in train_lines:
            frame_path, label = line.strip().rsplit(' ', 1)
            new_path = os.path.join(train_images_dir, os.path.basename(frame_path)).replace("\\", "/")
            copy2(frame_path, new_path)
            train_file.write(f"{new_path} {label}\n")
        for line in val_lines:
            frame_path, label = line.strip().rsplit(' ', 1)
            new_path = os.path.join(val_images_dir, os.path.basename(frame_path)).replace("\\", "/")
            copy2(frame_path, new_path)
            val_file.write(f"{new_path} {label}\n")

    print("Dataset split completed successfully!")

label_file = "label250.txt"  
output_dir = "250Split"
split_lf(label_file, output_dir, train_ratio=0.8)
```

# MobileNet Model Training
MobileNet is designed for mobile and embedded vision applications, making it lightweight. It uses depthwise separable convolutions, which significantly reduce the number of parameters and computational cost, while maintaining high accuracy.  
MobileNet is used for classification task (Violent or Non-Violent) in the project. The final classification layer is modified to suit binary classification.  

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from PIL import Image
from sklearn.metrics import precision_score, recall_score, f1_score
from tqdm import tqdm
```

```python
class CustomDataset(Dataset):
    def __init__(self, file_path, transform=None):
        self.data = []
        self.labels = []
        with open(file_path, 'r') as f:
            for line in f:
                parts = line.strip().split(' ')
                image_path = parts[0]
                label = int(parts[1])
                self.data.append(image_path)
                self.labels.append(label)
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = self.data[idx]
        label = self.labels[idx]
        img = Image.open(img_path).convert('RGB')  # RGB format
        if self.transform:
            img = self.transform(img)
        return img, label

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_dataset = CustomDataset(file_path='250Split/train250.txt', transform=transform)
val_dataset = CustomDataset(file_path='250Split/val250.txt', transform=transform)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
model = models.mobilenet_v2(pretrained=True)

# Replace the final classification layer with dropout included
model.classifier = nn.Sequential(
    nn.Dropout(0.3),  # Dropout layer with 30% probability
    nn.Linear(model.classifier[1].in_features, 2)  # Final classification layer
)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define loss function and optimizer with L2 regularization
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001, weight_decay=1e-4)  # Add weight decay

# Learning Rate Scheduler
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)

# Early stopping parameters
early_stopping_patience = 5
best_val_loss = float('inf')
early_stopping_counter = 0

# Gradient Clipping Threshold
max_grad_norm = 5.0

# Train the model
num_epochs = 25
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    with tqdm(train_loader, unit="batch") as tepoch:
        for inputs, labels in tepoch:
            tepoch.set_description(f"Epoch {epoch + 1}")
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            
            # Gradient Clipping
            nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)
            optimizer.step()
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            tepoch.set_postfix(loss=loss.item())
    epoch_loss = running_loss / len(train_loader)
    epoch_accuracy = 100 * correct / total
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")
    
    # Validation
    model.eval()
    val_loss = 0.0
    val_correct = 0
    val_total = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    val_loss /= len(val_loader)
    val_accuracy = 100 * val_correct / val_total
    precision = precision_score(all_labels, all_preds, average='binary')
    recall = recall_score(all_labels, all_preds, average='binary')
    f1 = f1_score(all_labels, all_preds, average='binary')
    
    print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.2f}%, Precision: {precision:.2f}, Recall: {recall:.2f}, F1 Score: {f1:.2f}")
    
    # Check for early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        early_stopping_counter = 0
        torch.save(model.state_dict(), 'best_model_optimized.pth')
    else:
        early_stopping_counter += 1
    
    if early_stopping_counter >= early_stopping_patience:
        print(f"Early stopping triggered after {epoch+1} epochs.")
        break
    
    # Step the scheduler
    scheduler.step()

print(f"Training completed. Best model saved to 'best_model_optimized.pth'.")
```

The trained model is tested on a separate dataset of 30 Violent and 30 Non-Violent videos and evaluated to calculate accuracy, precision, recall and f1 scores.

```python
import torch
from torchvision import models, transforms
from PIL import Image
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tqdm import tqdm
```

```python
class_names = ['NonViolence', 'Violence']
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = models.mobilenet_v2(pretrained=False)
model.classifier[1] = torch.nn.Linear(model.classifier[1].in_features, 2)
model.load_state_dict(torch.load('best_model_optimized.pth'))  # Load your trained model
model.to(device)
model.eval()
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def extract_frames(video_path, frame_skip=10):
    cap = cv2.VideoCapture(video_path)
    frames = []
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_skip == 0:
            frames.append(frame)
        frame_count += 1
    cap.release()
    return frames

import torch.nn.functional as F  # For softmax

# Predict class probabilities for a single frame
def predict_frame_probabilities(frame):
    image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)) 
    image = transform(image).unsqueeze(0).to(device) 
    with torch.no_grad():
        output = model(image)  # Raw logits
        probabilities = F.softmax(output, dim=1)  # Apply softmax to get probabilities
    return probabilities.cpu().numpy().squeeze()
    
def test_model_on_videos(test_videos_folder):
    y_true, y_pred, y_prob = [], [], [] 
    for label, subdir in enumerate(['NonViolence', 'Violence']):
        subdir_path = os.path.join(test_videos_folder, subdir)
        video_files = [f for f in os.listdir(subdir_path) if f.endswith(('.mp4', '.avi', '.mov'))]
        for video_file in tqdm(video_files, desc=f"Processing {subdir} videos"):
            video_path = os.path.join(subdir_path, video_file)
            frames = extract_frames(video_path, frame_skip=10) 
            frame_probs = [predict_frame_probabilities(frame) for frame in frames]  
            
            # Average probabilities for video-level prediction
            avg_probs = torch.mean(torch.tensor(frame_probs), dim=0).numpy()
            predicted_class = int(avg_probs[1] > avg_probs[0]) 
            y_true.append(label)
            y_pred.append(predicted_class)
            y_prob.append(avg_probs)
            print(f"Video: {video_file} -> Predicted: {class_names[predicted_class]}, "
                  f"True: {class_names[label]}")
    
    return y_true, y_pred, y_prob

test_videos_folder = 'Test_vids_2/'

y_true, y_pred, y_prob = test_model_on_videos(test_videos_folder)

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, PrecisionRecallDisplay
import numpy as np

# Plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap=plt.cm.Blues, values_format='d')
    plt.title("Confusion Matrix")
    plt.show()

# Plot bar chart for evaluation metrics
def plot_metrics(accuracy, precision, recall, f1):
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    values = [accuracy, precision, recall, f1]
    
    plt.bar(metrics, values, color=['#FFB6C1', '#87CEFA', '#98FF98', '#FFD700'])
    plt.ylim(0, 1)  # Since metrics are in the range [0, 1]
    plt.ylabel('Score')
    plt.title('Evaluation Metrics', pad=20)
    for i, v in enumerate(values):
        plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontsize=12)
    plt.show()

# Plot Precision-Recall curve
def plot_precision_recall_curve(y_true, y_pred):
    display = PrecisionRecallDisplay.from_predictions(y_true, y_pred, name="Model")
    display.ax_.set_title("Precision-Recall Curve")
    plt.show()

plot_confusion_matrix(y_true, y_pred, class_names)
plot_metrics(accuracy, precision, recall, f1)
plot_precision_recall_curve(y_true, y_pred)
```
